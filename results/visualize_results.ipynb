```python
import plotly.graph_objects as go

import numpy as np
import pandas as pd

from scipy import signal
```

```python
r_eval = pd.read_csv('aliens_3tl_eval_reward.csv', sep=',', header=0)
# r_eval = r_eval.drop(columns="Wall time")

r_train = pd.read_csv('aliens_3tl_mean_reward.csv', sep=',', header=0)
# r_train = r_train.drop(columns="Wall time")

r_eval_2 = pd.read_csv('aliens_1tl_eval_reward.csv', sep=',', header=0)
# r_eval_2 = r_eval.drop(columns="Wall time")

r_train_2 = pd.read_csv('aliens_1tl_mean_reward.csv', sep=',', header=0)
# r_train_2 = r_train.drop(columns="Wall time")
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x = r_eval['Step'],
    y = signal.savgol_filter(r_eval['Value'], 53, 4),
    # y = r_eval['Value'],
    mode='lines',
    line=go.scatter.Line(color="red"),
    name='Eval'
))

fig.add_trace(go.Scatter(
    x=r_train['Step'],
    y=signal.savgol_filter(r_train['Value'], 53, 4), 
    # y = r_train['Value'],
    mode='lines',
    line=go.scatter.Line(color="blue"),
    name='Train'
))

fig.add_trace(go.Scatter(
    x = r_eval_2['Step'],
    y = signal.savgol_filter(r_eval_2['Value'], 53, 4),
    # y = r_eval['Value'],
    mode='lines',
    line=go.scatter.Line(color="green"),
    name='Eval'
))

fig.add_trace(go.Scatter(
    x=r_train_2['Step'],
    y=signal.savgol_filter(r_train_2['Value'], 53, 4), 
    # y = r_train['Value'],
    mode='lines',
    line=go.scatter.Line(color="yellow"),
    name='Train'
))


fig.show()
```
